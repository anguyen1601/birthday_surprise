```python
import os
import instaloader

# 1. Configuration
# Your personal Instagram username (the account that follows the private profile)
MY_USERNAME = "YOUR_USERNAME"

# The full URL or just the shortcode of the private Reel
# Example URL: https://instagram.com -> shortcode is C8XyZ_uoxYz
REEL_URL_OR_SHORTCODE = "https://instagram.com"


def extract_shortcode(url_or_code):
    """Extracts the 11-character shortcode from a full URL if necessary."""
    if "instagram.com" in url_or_code:
        # Splits the URL by slashes and filters out empty spaces
        parts = [p for p in url_or_code.split("/") if p]
        # The shortcode usually follows 'p' or 'reel'
        if "reel" in parts:
            return parts[parts.index("reel") + 1]
        elif "p" in parts:
            return parts[parts.index("p") + 1]
    return url_or_code


def download_private_reel():
    # Initialize Instaloader instance
    L = instaloader.Instaloader()

    # Clean the shortcode from the input
    shortcode = extract_shortcode(REEL_URL_OR_SHORTCODE)
    print(f"Target Reel Shortcode: {shortcode}")

    try:
        # 2. Login Process
        # Try loading an existing session file first to avoid repeating logins
        session_file = os.path.expanduser(f"~/.instaloader-session-{MY_USERNAME}")
        if os.path.exists(session_file):
            print("Loading saved session...")
            L.load_session_from_file(MY_USERNAME)
        else:
            print(f"Logging in as {MY_USERNAME}...")
            # This will prompt you for your password securely in the terminal
            L.interactive_login(MY_USERNAME)
            L.save_session_to_file()

        # 3. Download the Reel
        print("Fetching post data...")
        post = instaloader.Post.from_shortcode(L.context, shortcode)

        print(f"Downloading Reel from private user: @{post.owner_username}")
        # Downloads video, thumbnail, and caption text into a folder named after the shortcode
        L.download_post(post, target=f"reel_{shortcode}")
        print(f"Success! Check the folder: 'reel_{shortcode}'")

    except instaloader.exceptions.BadCredentialsException:
        print("Error: Invalid username or password.")
    except instaloader.exceptions.LoginRequiredException:
        print("Error: You must be logged in to view this private content.")
    except instaloader.exceptions.ProfileNotExistsException:
        print("Error: The target profile or post shortcode does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


if __name__ == "__main__":
    download_private_reel()
```
